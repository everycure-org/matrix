```python
base_release_file_path = "gs://data.dev.everycure.org/data/01_RAW/drug_list/v0.2.0/ec-drug-list.parquet"
release_file_path = "../data/03_primary/release/ec-drug-list.parquet"
```

```python
import pandas as pd

raw_base_df = pd.read_parquet(base_release_file_path)
raw_release_df = pd.read_parquet(release_file_path)
```

```python
base_df = raw_base_df.copy()
release_df = raw_release_df.copy()
```

```python
column_to_compare = "translator_id"

base_df = base_df[["id", "name", column_to_compare]]
release_df = release_df[["id", column_to_compare]]
merged_df = base_df.merge(release_df, on="id", how="inner", suffixes=("_base", "_release"))
merged_df[merged_df[column_to_compare + "_base"] != merged_df[column_to_compare + "_release"]]

merged_df[merged_df[column_to_compare + "_base"] != merged_df[column_to_compare + "_release"]].to_csv(
    "change_in_translator_id.csv", index=False
)
```
